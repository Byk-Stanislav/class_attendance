```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
infa = pd.read_csv('testovoeee.csv')
```

```python
infa.isna().any().any()
```

```python
infa.dtypes
```

```python
infa['event_date'] = pd.to_datetime(infa['event_date'])
```

```python
df_for_dynamics_of_the_number_of_visits_by_groups = infa.groupby(['event_date', 'group_ids'])['is_attend'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_for_dynamics_of_the_number_of_visits_by_groups, x='event_date', y='is_attend', hue='group_ids', marker='o')
plt.title('    ')
plt.xlabel('')
plt.ylabel(' ')
plt.xticks(rotation=45)
plt.legend(title='')
plt.grid()
plt.show()
```

```python
df_for_teachers_attendance_by_day = infa.groupby(['event_date', 'teacher_ids'])['is_attend'].mean().unstack()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_for_teachers_attendance_by_day * 100, cmap='viridis', annot=True, fmt='.2f', cbar_kws={'label': '  '})
plt.title('       , %')
plt.xlabel('')
plt.ylabel(' ')
```

```python
df_for_groups_attendance_by_day = infa.groupby(['event_date', 'group_ids'])['is_attend'].mean().unstack()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_for_groups_attendance_by_day * 100, cmap='viridis', annot=True, fmt='.2f', cbar_kws={'label': '  '})
plt.title('       , %')
plt.xlabel('')
plt.ylabel(' ')
```
